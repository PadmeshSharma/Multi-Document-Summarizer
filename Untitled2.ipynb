punctuation removal  

```python
import csv
import string
exclude=string.punctuation

# Function to remove punctuation from a string
def remove_punctuation(text):
    translator = str.maketrans('', '', exclude)
    return text.translate(translator)
```

```python
df['article']=df['article'].apply(remove_punctuation)
df['highlights']=df['highlights'].apply(remove_punctuation)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
# Function to remove stop words from a string
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)
```

```python
df['article']=df['article'].apply(remove_stopwords)
df['highlights']=df['highlights'].apply(remove_stopwords)
```

sentence and word tokenization

```python
nltk.download('punkt')
# Function to tokenize sentences and words
def tokenize_data(text):
    sentences = nltk.sent_tokenize(text)
    words = [nltk.word_tokenize(sentence) for sentence in sentences]
    return ' '.join(words[0])
```

```python
df['article']=df['article'].apply(tokenize_data)
df['highlights']=df['highlights'].apply(tokenize_data)
```
